## Muestreo

Empezamos generadon la muestra segun la solicitud del 60% de proporción entre test y entreinamiento.

```python
from sklearn.model_selection import train_test_split
X = data.drop("Target", axis=1)
y = data["Target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

## Modelización con Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
modelo_fit_NB = gnb.fit(X_train, y_train)

y_predict_NB_training = modelo_fit_NB.predict_proba(X_train)
y_predict_NB_test = modelo_fit_NB.predict_proba(X_test)
```

## Curva ROC


```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc

# Desde datos train
fpr_train, tpr_train, _ = roc_curve(y_train, y_predict_NB_training[:,1])

# Desde datos test
fpr_test, tpr_test, _ = roc_curve(y_test, y_predict_NB_test[:,1])
```

### Grafico curv ROC segun datos de entreinamiento

```python
plt.figure(figsize=(4,4))
plt.plot(fpr_train, tpr_train, color='darkorange')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

### Grafico curv ROC segun datos de test

```python
plt.figure(figsize=(4,4))
plt.plot(fpr_test, tpr_test, color='green')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

## Grafico en conjunto

```python
plt.figure(figsize=(4,4))
plt.plot(fpr_train, tpr_train, color='darkorange', label='Entrenamiento')
plt.plot(fpr_test, tpr_test, color='green', label='Test')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()
```

## AUC

## AUC con datos de entreinamiento

```python
auc_roc_train = roc_auc_score(y_train, y_predict_NB_training[:,1])
auc_roc_train
```

## AUC con datos de test

```python
auc_roc_test = roc_auc_score(y_test, y_predict_NB_test[:,1])
auc_roc_test
```

## Script de Optimización Modelación


En esta parte se intentará modificar ciertos parametros del muestro para identificar si se obtiene una mejora.

```python
best_test_size =[]
for i in range(10, 90, 1):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i/100, random_state=42)
  modelo_fit_NB = gnb.fit(X_train, y_train)
  y_predict_NB_training = modelo_fit_NB.predict_proba(X_train)
  y_predict_NB_test = modelo_fit_NB.predict_proba(X_test)
  fpr_train, tpr_train, _ = roc_curve(y_train, y_predict_NB_training[:,1])
  fpr_test, tpr_test, _ = roc_curve(y_test, y_predict_NB_test[:,1])
  auc_roc_train = roc_auc_score(y_train, y_predict_NB_training[:,1])
  auc_roc_test = roc_auc_score(y_test, y_predict_NB_test[:,1])
  best_test_size.append([i,auc_roc_test])
```

```python
best_test_size_dt = pd.DataFrame(best_test_size, columns = ["test_size", "auc_roc_test"])
plt.plot(best_test_size_dt["test_size"], best_test_size_dt["auc_roc_test"])
plt.xlabel("test_size")
plt.ylabel("auc_roc_test")
plt.show()
```

```python
#El mejor rango de test_size es:
row=best_test_size_dt['auc_roc_test'].idxmax()
print("el tamaño que da el mejor resultado es con por lo menos ", 100-best_test_size_dt.loc[row]["test_size"], "% de training \n ya que se obtiene el mayor valor de auc score que es: ", best_test_size_dt.loc[row]["auc_roc_test"])
```
